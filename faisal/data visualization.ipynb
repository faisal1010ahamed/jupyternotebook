```python
import pandas as pd
```

```python
from sklearn.preprocessing import Normalizer
```

```python
normalization_df=pd.DataFrame({'x':np.random.randint(-100,100,1000),'y':np.random.randint(-80,80,1000),'z':np.random.randint(-150,150,1000)})
```

```python
normalization_df
```

```python
ax=plt.axes(projection='3d')
ax.scatter3D(normalization_df.x,normalization_df.y,normalization_df.z)
plt.figure()
```

```python
nol=Normalizer()
```

```python
normalization_df=nol.fit_transform(normalization_df)
```

```python
normalization_df
```

```python
norm_df=pd.DataFrame(normalization_df,columns=['x','y','z'])
```

```python
norm_df
```

```python
ax=plt.axes(projection='3d')
ax.scatter3D(norm_df.x,norm_df.y,norm_df.z)
plt.figure()
```

```python
data={'First Score':[100,75,np.nan,80],'Second Score':[np.nan,54,65,np.nan],'Third Score':[np.nan,50,60,100]}
```

```python
data
```

```python
var=pd.DataFrame(data)
```

```python
var
```

```python
var.isnull()
var.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer 
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
imputer.fit(var)
imputer.transform(var)
```
