```python
from bs4 import BeautifulSoup
```

```python
import requests 
```

```python
url="https://engineering.careers360.com/colleges/ranking/top-engineering-colleges-in-tamil-nadu"
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
colleges=soup.select("h2 > a")
```

```python
colleges
```

```python
Ratings=soup.select(".rateTag")
```

```python
Ratings
```

```python
import pandas as pd
```

```python
Collegeslist=pd.DataFrame(colleges,columns=["Names of college"])
```

```python
Collegeslist
```

```python
Collegeslist["Reviews"]=Ratings
```

```python
Ownership=soup.select(".middleText span+ a")
```

```python
Collegeslist["Ownership"]=Ownership
```
